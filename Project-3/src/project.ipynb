# Data Cleaning and Transformation
Implemented Requirements:
- Type conversion
- Transform a column 
- Create a new calculated column

I changed the rating column to a float and the rating_count column to an int when both were originally strings.

```python
df = df.astype({'rating': float, 'rating_count': int})
df
```

I transformed the values in the name, subhead, and description columns to uppercase strings so it's easier to use.

```python
df['name'] = df['name'].str.upper()
df['subhead'] = df['subhead'].str.upper()
df['description'] = df['description'].str.upper()
df
```

I created a new column that has the base flavor of each ice cream flavor (the actual flavor of the ice cream without the extra mix-ins). I got the base flavor by splitting the subheading, getting the part of the string right before 'ICE CREAM' and if necessary, right after 'OF'.

```python
df['base_flavor'] = df['subhead'].str.split(' ICE CREAM').str[0]
df['base_flavor'] = df['base_flavor'].str.split(' OF').str.get(-1)
df
```

# Descriptive Statistics
Implemented Requirements:
- Calculate summary statistics
- Calculate value counts

I used pandas methods to get these descriptive statistics.

```python
rating_mean = df['rating'].mean()
print('Mean: ' + str(rating_mean))
rating_median = df['rating'].median()
print('Median: ' + str(rating_median))
rating_25 = df['rating'].quantile(0.25)
print('25th Percentile: ' + str(rating_25))
rating_75 = df['rating'].quantile(0.75)
print('75th Percentile: ' + str(rating_75))
rating_min = df['rating'].min()
print('Min: ' + str(rating_min))
rating_max = df['rating'].max()
print('Max: ' + str(rating_max))
```

I found out what the lowest rated flavor(s) was/were by finding the row that had the value in the 'rating' column equal to the minimum rating.

```python
print('Lowest Rated Flavor(s)')
df.loc[df['rating'] == rating_min]
```

The lowest rated flavor is Chocolate Chip Cookie Dough Core with a rating of 1.8.

I found out what the highest rated flavor(s) was/were by finding the row that had the value in the 'rating' column equal to the maximum rating.

```python
print('Highest Rated Flavor(s)')
df.loc[df['rating'] == rating_max]
```

The highest rated flavors are Chocolate Peanut Butter Split and Ice Cream Sammie with ratings of 5.0.

I got the mode using value counts on the 'rating' column.

```python
df['rating'].value_counts()
```

The mode amongst the ratings is 4.7. 12 flavors were given a rating of 4.7.

# Analysis on Categorical Data

Slightly different from the question I answered in Homework 02, I found how many flavors used chocolate as a base flavor (as opposed to just had chocolate as a main ingredient). I did this by going through the base flavor column I created and summing up how many rows mentioned the word 'CHOCOLATE'.

```python
chocolate_count = df['base_flavor'].str.contains('CHOCOLATE').sum()
print('Flavors with Chocolate as a Base Flavor: ' + str(chocolate_count))
```

To begin finding the common ingredients between the top 5 rated flavors, I first created a new DataFrame with just the top 5 rated flavors. I first organized the flavors by rating, then by rating count (as in, if any flavors had the same rating, the flavor that's been rated more often will take precedence). I then used the head method and passed 5 in to get just the top 5 rated flavors.

```python
df_sorted = df.sort_values(['rating', 'rating_count'], ascending=[False, False])
top_5 = df_sorted.head(5)
top_5
```

I found the common ingredients by getting the string of ingredients of the top rated flavor (Ice Cream Sammie) and then turning it into a list. I then iterated through this list and then for each ingredient in the list, summed up how many rows in the ingredients column mentioned said ingredient. If this value was equal to 5 (which means its in every single flavor), I printed said ingredient.

```python
import re

top_flavor_ingredients_str = top_5.loc[top_5['name'] == 'ICE CREAM SAMMIE', 'ingredients'].item()
top_flavor_ingredients = re.split(r', \s*(?![^()]*\))', top_flavor_ingredients_str)
print('Common Ingredients In Top 5 Flavors')
for ingredient in top_flavor_ingredients:
    if '(' in ingredient:
            ingredient = ingredient[:(ingredient.index('(')-1)]
    if top_5['ingredients'].str.contains(ingredient).sum() == 5:
        print(ingredient)
```

# Visualizations
Implemented Requirements:

This visualization plots ratings in ascending order versus their rating count using a line graph. I used the DataFrame I created where I sorted the flavors by rating and rating count.

```python
import matplotlib.pyplot as plt

plt.plot(df_sorted['rating'], df_sorted['rating_count'])
plt.xlabel('Rating')
plt.ylabel('Rating Count')
plt.show()
```

This visualization plots each ingredient versus how many flavors it's used in using a horizontal bar graph. I did the same exact thing I did in Homework 02 except this time, I iterated through the original DataFrame (df) and looked at the ingredients column.

```python
ingredients_count_all = {}
for ind in df.index:
    ingredients = re.split(r', \s*(?![^()]*\))', df['ingredients'][ind])
    for ingredient in ingredients:
        if '(' in ingredient:
            ingredient = ingredient[:(ingredient.index('(')-1)]
        elif '.' in ingredient:
            ingredient = ingredient[:(ingredient.index('.'))]
        if ingredient in ingredients_count_all:
            ingredients_count_all[ingredient] += 1
        else:
            ingredients_count_all[ingredient] = 1
ingredients = list(ingredients_count_all.keys())
ingredient_counts = list(ingredients_count_all.values())
plt.barh(ingredients, ingredient_counts)
plt.title("How Often Ingredients Are Used In Ben & Jerry's Ice Cream Flavors")
plt.ylabel('Ingredient')
plt.xlabel("How Many Flavors It's Used In")
plt.rcParams['figure.figsize'] = [30, 40]
plt.show()
```
