## 4. Initial column (or row) clean-up

The ease-ment column should be removed because every value in that column is missing.

```python
df['EASE-MENT'].notna().sum()
```

```python
df = df.drop('EASE-MENT', 1)
df
```

Every row where every column value is missing should be removed.

```python
df.isna().all(axis=1).sum()
```

```python
df = df.dropna(how = 'all')
```

I transformed the columns I listed above into integers because they were either floats that never used a decimal or a string with commas that also never used a decimal. Making all these columns integers make them easier to use. Additionally, I made sure all leading and trailing white space is removed from each column name.

```python
df = df.rename(columns=lambda x: x.strip())
df = df.astype({'BOROUGH': 'Int64', 'BLOCK': 'Int64', 'LOT': 'Int64', 'ZIP CODE': 'Int64', 'RESIDENTIAL UNITS': 'Int64', 
                'COMMERCIAL UNITS': 'Int64', 'TOTAL UNITS': 'Int64', 'YEAR BUILT': 'Int64', 'TAX CLASS AT TIME OF SALE': 'Int64'})
df['LAND SQUARE FEET'] = df['LAND SQUARE FEET'].fillna('0')
df['GROSS SQUARE FEET'] = df['GROSS SQUARE FEET'].fillna('0')
df['SALE PRICE'] = df['SALE PRICE'].fillna('0')
df = df.astype({'LAND SQUARE FEET': str, 'GROSS SQUARE FEET': str, 'SALE PRICE': str})
df['LAND SQUARE FEET'] = df['LAND SQUARE FEET'].replace(',', '', regex=True)
df['GROSS SQUARE FEET'] = df['GROSS SQUARE FEET'].replace(',', '', regex=True)
df['SALE PRICE'] = df['SALE PRICE'].replace(',', '', regex=True)
df = df.astype({'LAND SQUARE FEET': int, 'GROSS SQUARE FEET': int, 'SALE PRICE': int})
df
```

## 6. Describe the kind of buildings that were sold

To initially clean the data, I dropped every row value in the 'BUILDING CLASS AT PRESENT' column that was missing. I then found out how much of each building was sold (one family homes, office buildings, condominiums) by summing up the values in the column that contained each type of building's corresponding letter based on their general class of properties ("A" signifies one-family homes, "O" signifies office buildings, and "R" signifies condominiums). This is because each value in the 'BUILDING CLASS AT PRESENT' column is a string of two characters, the first being a letter that corresponds to each building's genral class of properties. I then subtracted these three values from the total number of properties listed to get the number of other types of buildings.

```python
building_class_at_present = df['BUILDING CLASS AT PRESENT'].dropna()
one_family_homes_count = building_class_at_present.str.contains('A').sum()
office_buildings_count = building_class_at_present.str.contains('O').sum()
condominiums_count = building_class_at_present.str.contains('R').sum()
print('One Family Homes Count: ' + str(one_family_homes_count))
print('Office Buildings Countt: ' + str(office_buildings_count))
print('Condominiums Count: ' + str(condominiums_count))
number_of_rows = len(df.index)
other_count = number_of_rows - one_family_homes_count - office_buildings_count - condominiums_count
print('Other Count: ' + str(other_count))
```

I used a horizontal bar graph to visualize the comparison of the number of the kinds of units sold. I used matplotlib to do this.

```python
import matplotlib.pyplot as plt

building_classes = ['One Family Homes', 'Office Buildings', 'Condominiums', 'Other']
unit_counts = [one_family_homes_count, office_buildings_count, condominiums_count, other_count]
plt.barh(building_classes, unit_counts)
plt.title('Number Of Units Sold for Each Building Type')
plt.ylabel('Building Class At Present')
plt.xlabel('Number Of Units Sold')
plt.show()
```

# 7. Calculate summary statistics for the prices of properties sold for all of Manhattan and for a couple of select neighborhoods

I used the DataFrame methods mean, median, quantile, max, and min to find the summary statistics using the 'SALE PRICE' column.

```python
print('Statistics For All Properties')
all_mean = df['SALE PRICE'].mean()
print('Mean: ' + str(all_mean))
all_median = df['SALE PRICE'].median()
print('Median: ' + str(all_median))
all_25 = df['SALE PRICE'].quantile(0.25)
print('25th Percentile: ' + str(all_25))
all_75 = df['SALE PRICE'].quantile(0.75)
print('75th Percentile: ' + str(all_75))
all_max = df['SALE PRICE'].max()
print('Max: ' + str(all_max))
all_min = df['SALE PRICE'].min()
print('Min: ' + str(all_min))
```

For my first neighborhood, I picked to find statistics for the Chinatown neighborhood (as that's where I dorm). I used the same DataFrame methods, however, I now also used the loc method to make sure I only considered rows where the 'NEIGHBORHOOD' column was equal to 'CHINATOWN' to only consider properties in Chinatown.

```python
print('Statistics For Chinatown Properties')
chinatown_mean = df.loc[df['NEIGHBORHOOD'] == 'CHINATOWN', 'SALE PRICE'].mean()
print('Mean: ' + str(chinatown_mean))
chinatown_median = df.loc[df['NEIGHBORHOOD'] == 'CHINATOWN', 'SALE PRICE'].median()
print('Median: ' + str(chinatown_median))
chinatown_25 = df.loc[df['NEIGHBORHOOD'] == 'CHINATOWN', 'SALE PRICE'].quantile(0.25)
print('25th Percentile: ' + str(chinatown_25))
chinatown_75 = df.loc[df['NEIGHBORHOOD'] == 'CHINATOWN', 'SALE PRICE'].quantile(0.75)
print('75th Percentile: ' + str(chinatown_75))
chinatown_max = df.loc[df['NEIGHBORHOOD'] == 'CHINATOWN', 'SALE PRICE'].max()
print('Max: ' + str(chinatown_max))
chinatown_min = df.loc[df['NEIGHBORHOOD'] == 'CHINATOWN', 'SALE PRICE'].min()
print('Min: ' + str(chinatown_min))
```

# 8. Bin the prices of properties sold

I created 5 bins based on the overall min to the Chinatown 25th percentile, the Chinatown 25th percentile to the overall median, the overall median to the Soho median, the Soho median to the overall mean, and the overall mean to the overall max.

```python
bins = [all_min-1, chinatown_25, all_median, soho_median, all_mean, all_max]
labels = ['Low', 'Mid-Low', 'Mid', 'Mid-High', 'High']
price_bins = pd.cut(df['SALE PRICE'], bins, labels=labels)
price_bins_counts = pd.value_counts(price_bins, sort=False)
print(price_bins_counts)
```

I visualized the values in each bin through a bar graph.

```python
price_bins_plot = price_bins_counts.plot.bar()
```

# 10. Which month did the least amount of sales occur

I used the calendar module in this step. I first created a new DataFrame with just the sale data column, splitting it by the '/' character to split the month, day, and year with the month being in the first column (column 0). I then used the calendar month_abbr method to change all the months values (which were previously shown in numbers ex. January = 1) to their string abbreviations. I then used value counts to sum up how often each month is mentioned in the DataFrame to get how many properties were sold each month.

```python
import calendar
    
dates_split = df['SALE DATE'].str.split('/', expand=True)
dates_split[0] = dates_split[0].apply(lambda x: calendar.month_abbr[int(x)])
dates_split[0].value_counts()
```

```python
months_plot = dates_split[0].value_counts().plot.bar()
```

The least number of sales occured in August. Given that the month with the most number of sales is June, which is the beginning of the summer (which is when it's warmer and more people are free so it's an optimal time to move), the reason the least number of sales occurs in August is probably because it's the end of the summer when most people who are looking to move around summer time have already done so.
