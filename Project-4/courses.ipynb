```python
import re
import pandas as pd
from bs4 import BeautifulSoup
```

# Spring 2021 Courses

```python
sc = BeautifulSoup(open("springcourses.html"), "html.parser")
```

```python
rows = sc.select('ul.schedule-listing .row')
```

```python
springcourses = []
for row in rows:

    parts = row.select('span')
        
    course = []

    numbersection = parts[0].text.strip().split('\n')[0]
    if '(' in numbersection:
        course.append(numbersection[:numbersection.index('(')])
    else:
         course.append(numbersection)

    name = parts[1].text.strip().split('\n')
    name = [s.strip() for s in name]
    name = list(filter(None, name))
    if len(name) == 2:
        name = name[0] + ' ' + name[1]
    else:
        name = name[0]
    course.append(name)
    
    instructors = parts[2].text
    if 'Office Hours' in instructors:
        instructors = instructors[:instructors.index('Office Hours')]
    instructors = instructors.strip().split('\n')
    instructors = [instructor.strip() for instructor in instructors]
    instructors = list(filter(None, instructors))
    instructorsStr = ''
    for instructor in instructors:
        instructorsStr += instructor + ', '
    course.append(instructorsStr[:len(instructorsStr)-2])
    
    time = parts[3].text.strip()
    course.append(time)
    
    springcourses.append(course)
```

```python
springcoursesdf = pd.DataFrame(springcourses, columns = ['Number-Section', 'Name', 'Instructor', 'Time'])
```

```python
springcoursesdf
```

```python
springcoursesdf['Number'] = springcoursesdf['Number-Section'].str.extract('(\w*-\w*.\d{4})')
springcoursesdf['Section'] = springcoursesdf['Number-Section'].str.extract('-(.\d{3})')
springcoursesdf
```

# Course Catalog

```python
cc = BeautifulSoup(open("coursecatalog.html"), "html.parser")
```

```python
rows2 = cc.select('ul.courses-listing .col-sm-12')
```

```python
courses = []
for row in rows2:

    parts = row.select('p')
        
    course = []

    number = parts[0].text.strip()
    course.append(number)

    prereqs = parts[2].text.strip()
    course.append(prereqs)

    points = parts[1].text.strip()
    course.append(points)
    
    courses.append(course)
```

```python
coursesdf = pd.DataFrame(courses, columns = ['Number', 'Prereqs', 'Points'])
```

```python
coursesdf
```

```python
coursesdf['Number'] = coursesdf['Number'].str.extract('(\w*-\w*.\d{4})')
coursesdf['Prereqs'] = coursesdf['Prereqs'].str.extract(': (.*)')
coursesdf['Points'] = coursesdf['Points'].str.extract('(.*) Points')
coursesdf
```

```python
coursesdf.info()
```

```python
combinedcourses = pd.merge(springcoursesdf, coursesdf, on='Number', how='left')
```

I then stripped each element in the list and took out all the empty strings. For the courses cross listed in other departments, I decided to just include their number-section as listed in the CS department. I did this by first by after getting the text from the number-section part, stripping it and splitting it by the newline character into a list. I then only used the string in the first index (0) of the list (as this would be the number-section listed for the CS department. Also, as all courseshad a number in parentheses listed under it (I'm not sure what that represents but I know it's not needed) I would make sure to remove this by slicing the string if necessary to remove the part after the '(' inclusive. For the instructors with the "Office Hours" link under their name, I would slice the string to just get the part before the "Office Hours" substring. Finally, if a course had multiple instructors, as this also messed with formatting, after removing the "Office Hours" substring, I stripped and splitted by the newline character into a list. I then stripped each element in the list and took out all the empty strings. Finally, I iterated over the list, adding each instructor into a new string with commas after each instructor except for the last one which I did through string slicing at the end.

Some issues I saw in the Course Catalog page were that you couldn't get any of the needed categories (number, prerequisites, and points) alone. I used regex extracting to only get the parts I needed (as in only getting the number without the course name for number, removing the "Prerequisites: " substring at the start for rerequisites, and only getting the numbers for points). An additional issue I saw in the Course Catalog is not all the points for courses were just individual numbers. For Internship and Independent Study, points ranged from 1-4. To deal with this, I kept the points columns as type string instead of making them ints.

A final issue I noticed (but did not fix) is that as the Spring Courses include recitation sections for courses which have the same number but different section number as their respective classes, when you combined the Spring Courses DataFrame and the Course Catalog DataFrame by the 'Number' column, both the courses and their recitation sections will show up in the final DataFrame showing their respective different names, instructors, and times, but the same repeated prerequisites and points (even though you can't exactly have prerequisites and points for a recitation section as it is an extension of a class). I would fix this by maybe replacing the string in the preqresuities column to say "Must be enrolled in (insert class name here)." and replacing the number(s) in the points column with "-" since recitation sections themselves technically can't award points.

Using how=left allows for only the rows in the Course Catalog DataFrame which have a match in the 'Number' column in the Spring Course DataFrame to be added in. So in the final DataFrame, all Spring Courses have their 'Number', 'Name', 'Instructor', and 'Time' column information from the Spring Course DataFrame and their respective 'Prereqs' and 'Points' column information from the Course Catalog DataFrame. Any row with a course number shown in the Course Catalog DataFrame that does not have a matching course number in the Spring Courses DataFrame does not show up in the final DataFrame.
